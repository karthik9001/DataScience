# Hierachical TimeSeries Forecasting, Using Facebook Prophet Library
Warehouse > Product_Category > Product_Code

```python
import pandas as pd
df =pd.read_csv('data/Retail_Dataset2.csv',engine='python')
df.head()
```

```python
product_code_list=df.Product_Code.unique()
product_code_list.sort()
```

```python
Product_Category_list=df.Product_Category.unique()
Product_Category_list.sort()
```

```python
Warehouse_list=df.Warehouse.unique()
Warehouse_list.sort()
```

```python
print(Warehouse_list)
```

```python
from fbprophet import Prophet
import datetime
from datetime import datetime,timedelta
from pandas import DataFrame
from pandas import to_datetime
import pandas as pd

def prophet_prediction_static(df,warehouse, p_category, product, predict_range):
    new_df=df[(df.Warehouse == warehouse)&(df.Product_Category == p_category)&(df.Product_Code == product)].sort_values(by='Date')
    new_df['Date'] = pd.to_datetime(new_df['Date']) - pd.to_timedelta(7, unit='d')
    new_df=new_df.groupby([pd.Grouper(key='Date', freq='W')])['Order_Demand'].sum()
    new_df=new_df.to_frame()
    new_df.reset_index(inplace=True)
    model = Prophet()
    new_df=new_df.rename(columns={"Date": "ds", "Order_Demand": "y"})
    model.fit(new_df)
    date=str(new_df['ds'].max())
    #print(date[0:10])
    future = list()
    date1 = datetime.strptime(date[0:10], "%Y-%m-%d")
    for i in range(0,predict_range):
        date1 = date1 + timedelta(days=7)
        future.append([date1])
    future = DataFrame(future)
    future.columns = ['ds']
    future['ds']= to_datetime(future['ds'])
    forecast = model.predict(future)
    print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
    model.plot(forecast);
    
    
```

```python
df =pd.read_csv('data/Retail_Dataset2.csv',engine='python')
prophet_prediction_static(df,warehouse="Whse_S", p_category="Category_005", product="Product_0033",predict_range=10)
```

```python
def prophet_prediction_dynamic(warehouse, p_category, product, predict_range):
    df =pd.read_csv('data/Retail_Dataset2.csv',engine='python')
    new_df=df[(df.Warehouse == warehouse)&(df.Product_Category == p_category)&(df.Product_Code == product)].sort_values(by='Date')
    new_df['Date'] = pd.to_datetime(new_df['Date']) - pd.to_timedelta(7, unit='d')
    new_df=new_df.groupby([pd.Grouper(key='Date', freq='W')])['Order_Demand'].sum()
    new_df=new_df.to_frame()
    new_df.reset_index(inplace=True)
    #model = Prophet(yearly_seasonality = True)
    model = Prophet()
    new_df=new_df.rename(columns={"Date": "ds", "Order_Demand": "y"})
    model.fit(new_df)
    date=str(new_df['ds'].max())
    #print(date[0:10])
    future = list()
    date1 = datetime.strptime(date[0:10], "%Y-%m-%d")
    for i in range(0,predict_range):
        date1 = date1 + timedelta(days=7)
        future.append([date1])
    future = DataFrame(future)
    future.columns = ['ds']
    future['ds']= to_datetime(future['ds'])
    forecast = model.predict(future)
    print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']])
    model.plot(forecast);
```

# LSTM Model / RNN

```python

df =pd.read_csv('data/Retail_Dataset2.csv',engine='python')
new_df=df[(df.Warehouse == "Whse_J")&(df.Product_Category == "Category_001")&(df.Product_Code == "Product_0608")].sort_values(by='Date')
new_df['Date'] = pd.to_datetime(new_df['Date']) - pd.to_timedelta(7, unit='d')
new_df=new_df.groupby([pd.Grouper(key='Date', freq='W')])['Order_Demand'].sum()
new_df=new_df.to_frame()
new_df.reset_index(inplace=True)
new_df = new_df.set_index("Date")
new_df.head()
```

```python
#new_df.index.freq = 'MS'
ax = new_df['Order_Demand'].plot(figsize = (16,5), title = "Order_Demand")
ax.set(xlabel='Dates', ylabel='Order_Demand');
```

```python
train_data = new_df[:len(new_df)-5]
test_data = new_df[len(new_df)-5:]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(train_data)
scaled_train_data = scaler.transform(train_data)
scaled_test_data = scaler.transform(test_data)
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator

n_input = 12
n_features= 1
generator = TimeseriesGenerator(scaled_train_data, scaled_train_data, length=n_input, batch_size=1)
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM

lstm_model = Sequential()
lstm_model.add(LSTM(200, activation='relu', input_shape=(n_input, n_features)))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mse')

lstm_model.summary()
```

```python
lstm_model.fit_generator(generator,epochs=20)
```

```python
losses_lstm = lstm_model.history.history['loss']
plt.figure(figsize=(12,4))
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks(np.arange(0,21,1))
plt.plot(range(len(losses_lstm)),losses_lstm);
```

```python
lstm_predictions_scaled = list()

batch = scaled_train_data[-n_input:]
current_batch = batch.reshape((1, n_input, n_features))

for i in range(len(test_data)):   
    lstm_pred = lstm_model.predict(current_batch)[0]
    lstm_predictions_scaled.append(lstm_pred) 
    current_batch = np.append(current_batch[:,1:,:],[[lstm_pred]],axis=1)
```

```python
lstm_predictions_scaled
```

```python
lstm_predictions = scaler.inverse_transform(lstm_predictions_scaled)
lstm_predictions
```

```python
test_data['LSTM_Predictions'] = lstm_predictions
test_data
```

```python
test_data['Order_Demand'].plot(figsize = (16,5), legend=True)
test_data['LSTM_Predictions'].plot(legend = True);
```

```python
from statsmodels.tools.eval_measures import rmse
lstm_rmse_error = rmse(test_data['Order_Demand'], test_data["LSTM_Predictions"])
lstm_mse_error = lstm_rmse_error**2
mean_value = df['Order_Demand'].mean()

print(f'MSE Error: {lstm_mse_error}\nRMSE Error: {lstm_rmse_error}\nMean: {mean_value}')
```

# XGBoost Model

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import plot_importance, plot_tree
from sklearn.metrics import mean_squared_error, mean_absolute_error
plt.style.use('fivethirtyeight')
```

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index
    """
    df['Date'] = df.index
    df['hour'] = df['Date'].dt.hour
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['quarter'] = df['Date'].dt.quarter
    df['month'] = df['Date'].dt.month
    df['year'] = df['Date'].dt.year
    df['dayofyear'] = df['Date'].dt.dayofyear
    df['dayofmonth'] = df['Date'].dt.day
    df['weekofyear'] = df['Date'].dt.weekofyear
    
    X = df[['dayofweek','quarter','month','year',
           'dayofyear','dayofmonth','weekofyear']]
    if label:
        y = df[label]
        return X, y
    return X

#create_features(new_df, label="Order_Demand")

X_train, y_train = create_features(new_df, label='Order_Demand')
X_test, y_test = create_features(new_df, label='Order_Demand')
```

```python
reg = xgb.XGBRegressor(n_estimators=1000)
reg.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        early_stopping_rounds=50,
       verbose=False) # Change verbose to True if you want to see it train
```

```python
_ = plot_importance(reg, height=0.9)
```

```python
test_data["XGBoost_Predictions"]=reg.predict(X_test)[-5:]
```

```python
XGBoost_rmse_error=rmse(y_test.iloc[len(y_test)-5:], reg.predict(X_test)[-5:])
XGBoost_mse_error=rmse(y_test.iloc[len(y_test)-5:], reg.predict(X_test)[-5:])**2
```

```python
rmse_errors = [arima_rmse_error, lstm_rmse_error, prophet_rmse_error,XGBoost_rmse_error]
mse_errors = [arima_mse_error, lstm_mse_error, prophet_mse_error,XGBoost_mse_error]
errors = pd.DataFrame({"Models" : ["ARIMA", "LSTM", "Prophet","XGBoost"],"RMSE Errors" : rmse_errors, "MSE Errors" : mse_errors})
plt.figure(figsize=(16,9))
plt.plot_date(test_data.index, test_data["Order_Demand"], linestyle="-")
plt.plot_date(test_data.index, test_data["ARIMA_Predictions"], linestyle="-.")
plt.plot_date(test_data.index, test_data["LSTM_Predictions"], linestyle="--")
plt.plot_date(test_data.index, test_data["Prophet_Predictions"], linestyle=":")
plt.plot_date(test_data.index, test_data["XGBoost_Predictions"], linestyle=" ")
plt.legend()
plt.show()
```
